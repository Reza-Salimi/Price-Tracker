```python
import pandas as pd
import numpy as np
df = pd.read_excel(r"C:\Users\123\Downloads\mysql\price-history-1390-10-01-to-1399-10-01.foolad.xlsx", header=7, parse_dates=True)
dg = pd.read_excel(r"C:\Users\123\Downloads\mysql\price-history-1390-10-01-to-1399-10-01.ketram.xlsx", header=7, parse_dates=True)
```

```python
dg = dg[['پایانی','تاریخ میلادی']]
dg.set_index('تاریخ میلادی', inplace=True, drop=True)
```
